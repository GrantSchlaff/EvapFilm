```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Given data (t and z values) - swapped order
t = np.array([200.0, 200.0, 200.0, 240.0, 400.0, 150.0, 483.4, 233.4, 333.4, 233.4, 716.8, 250.0, 316.7, 316.7, 333.4,
               733.5, 866.84, 400.1, 266.7, 383.4, 500.1, 250.0, 216.7, 133.36, 400.1, 400.1, 566.8, 766.8, 416.8, 416.8, 183.37, 616.8, 
                              400.1, 466.8, 333.4, 383.4, 416.75, 483.43, 300.1, 200.0, 416.8, 333.4, 433.4, 133.4, 333.4])

z = np.array([7, 7, 8, 7, 25, 23, 30, 18, 24, 20, 37, 22, 13, 27, 25, 35, 30, 31, 17, 20, 25, 
             13, 15, 12, 22, 30, 17, 30, 23, 23, 15, 20, 20, 27, 23, 24, 26, 27, 25, 15, 22, 30, 25, 10, 28])

# Compute z/t ratio
ratio = z / t

# Generate probability density function (PDF)
x = np.linspace(min(ratio), max(ratio), 100)
kde = stats.gaussian_kde(ratio)
pdf_values = kde(x)

# Plot the PDF with shading
plt.figure(figsize=(8, 6))
plt.plot(x, pdf_values, label='PDF of $z/t$', color='blue')
plt.fill_between(x, pdf_values, color='blue', alpha=0.3)  # Shaded area
plt.xlabel(r'Ratio $z/t$ (mm/ms)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.title('Probability Density Function of $z/t$', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

# Save the plot as a PDF
plt.savefig("zt_ratio_pdf_plot.pdf")

# Show the plot
plt.show()
```
